# Impact of Toss Outcome on the Match

```python
toss_win_match_win = (df['Toss_winner'] == df['Winner']).sum()
total_matches = len(df)
labels= ["Won After Toss", "Lost After Toss"]
values = [toss_win_match_win, total_matches - toss_win_match_win]
fig = px.pie(
    names= labels,
    values = values,
    title = ("Impact of Toss Match Outcome")
)
fig.show()
```

# Batting first vs Chasing

```python
df['First_Innings_Win'] = df['Score_of_first_innings'] > df['Score_of_second_innings']

bat_first_wins = df[df['First_Innings_Win']]['Winner'].count()
chasing_wins = df[~df['First_Innings_Win']]['Winner'].count()

print("Bat First Wins:", bat_first_wins)
print("Chasing Wins:", chasing_wins)
```

```python
fig= px.bar(
    x= ["Bat First", "Chasing"],
    y= [bat_first_wins, chasing_wins],
    title= "Batting Fist VS Chasing Wins"

)
fig.show()
```

## T20 World Cup 2022 Data Analysis: Key Insights

After the analysis performed in this notebook, here are some key insights into the T20 World Cup 2022:

### 1. Matches Played Per Team
*   Teams like Ireland, Sri Lanka, and Zimbabwe played the most matches (8 each), indicating their participation in the initial 'First Round' stage, in addition to the 'Group' stage.
*   Major teams like India, England, and Pakistan played 6-7 matches, reaching the later stages of the tournament.

### 2. Matches Won by Each Team
*   England, the eventual champions, secured the most wins (5 matches).
*   Netherlands, India, Sri Lanka, and Pakistan also showed strong performances with 4 wins each.
*   Interestingly, 'No Result' accounted for 4 instances, suggesting rain-affected or abandoned matches during the tournament.

### 3. Impact of Toss Outcome on Match Result
*   The analysis shows that in **16 out of 45 matches**, the team that won the toss also won the match. Conversely, in 29 matches, the team that won the toss lost the match. This suggests that winning the toss does not guarantee a win, as the 'Lost After Toss' category has a higher count.

### 4. Matches Won at Different Stadiums
*   The graphs depicting wins per stadium were not detailed enough to extract specific trends, but they show the distribution of wins across various venues, with some stadiums hosting more matches and thus having more wins recorded.

### 5. Top Scorers / Player of the Match
*   The `Player_of_the_match` data shows individual performances that were crucial for their team's victories. While a specific breakdown of top scorers was not explicitly calculated as a continuous value, the plot highlights which players received the award and for which winning team.

### 6. First & Second Innings Score Comparison
*   The scatter plot between `Score_of_first_innings` and `Score_of_second_innings` with the `Winner` overlay indicates that there isn't a direct linear relationship, but it does show various score combinations leading to wins. Teams could win by scoring high in the first innings and defending, or by chasing effectively.

### 7. Batting First vs Chasing
*   **Batting first resulted in 24 wins**, while **chasing resulted in 18 wins**. This indicates a slight advantage for teams batting first in the tournament, though the difference is not overwhelmingly large, suggesting a fairly balanced competition between setting a target and chasing.

### 8. Runs Scored in the 1st and 2nd Innings
*   The histograms show the distribution of scores. The `Score_of_first_innings` histogram reveals a spread of scores, with a notable peak around the 150-180 mark. The `Score_of_second_innings` histogram similarly shows a distribution, with many successful chases falling within a similar range.

### 9. Group Stage Wins
*   The 'Wins by Stage' bar chart illustrates how teams performed in different stages (First Round, Group, Semi Final, Final). England's sole win in the 'Final' stage, and one win in the 'Semi Final' stage, along with Pakistan's, highlights their progression to the tournament's climax. The 'Group' stage had the most diverse winners.

### 10. Score Correlation Heatmap
*   There is a positive correlation of `0.336455` between `Score_of_first_innings` and `Score_of_second_innings`. This weak positive correlation suggests that while higher first innings scores might be somewhat associated with higher second innings scores (or higher targets leading to higher chase attempts), it's not a strong dependency.

### 11. Win Margin Distribution
*   The most common win margins were '5 Wickets' (7 times), followed by 'No Result' (4 times), and '6 Wickets' (3 times). This highlights that many matches were won by chasing a target, and many were closely contested.

### 12. Highest Scores
*   The **highest score in the first innings was 205**.
*   The **highest score in the second innings was 180**.
